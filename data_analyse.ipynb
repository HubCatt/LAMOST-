```python
import numpy as np
import pandas as pd

result =pd.DataFrame(columns=('obsid','label','pred','error'))
file = open(r"C:/Users/53546/Desktop/error.txt") 
for line in file:
    name = float(line.split("\t")[0])
    label= float(line.split('\t')[1].split(':')[1])
    pred = float(line.split('\t')[2].split(':')[1])
    error = pred - label
    result=result.append(pd.DataFrame({'obsid':[name],'label':[label],'pred':[pred],'error':[error]}),ignore_index=True)
file.close()
result.to_csv('error_data.csv',index=0)
```

```python
result['obsid'].to_csv('name.csv',index=0)
```

```python
information = pd.read_csv(r'D:\PycharmProjects\DataProcess\4类数据集/stellar_data_catalog.csv')

for i in range(len(result)):
   result.loc[i,'teff'] = information[information['obsid']==result.loc[i,'obsid']]['teff'].to_numpy()
   result.loc[i,'logg'] = information[information['obsid']==result.loc[i,'obsid']]['logg'].to_numpy()
   result.loc[i,'feh'] = information[information['obsid']==result.loc[i,'obsid']]['feh'].to_numpy()
   result.loc[i,'snru'] = information[information['obsid']==result.loc[i,'obsid']]['snru'].to_numpy()
   result.loc[i,'snrg'] = information[information['obsid']==result.loc[i,'obsid']]['snrg'].to_numpy()
   result.loc[i,'snrr'] = information[information['obsid']==result.loc[i,'obsid']]['snrr'].to_numpy()
   result.loc[i,'rv'] = information[information['obsid']==result.loc[i,'obsid']]['rv'].to_numpy()
   result.loc[i,'z'] = information[information['obsid']==result.loc[i,'obsid']]['z'].to_numpy()
result.to_csv('error_data.csv',index=0)
```
